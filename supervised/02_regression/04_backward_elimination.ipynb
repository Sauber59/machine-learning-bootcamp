### <a name='0'></a> Import bibliotek

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

sns.set()
np.random.seed(42)
np.set_printoptions(precision=4, suppress=True)
sklearn.__version__
````

### <a name='1'></a> Załadowanie danych

````python
df_raw = pd.read_csv('https://storage.googleapis.com/esmartdata-courses-files/ml-course/insurance.csv')
df_raw.head()
````

##Utworzenie kopii danych

````python
df = df_raw.copy()
df.info()
````

````python
df[df.duplicated()] #sprawdzamy czy sa duplikaty- to zle wplywa na wynik modelu
````

````python
df = df.drop_duplicates() #usuniecie duplikatow
df.info()
````

````python
##zamiana kolumn objektowych na kategoryczne np. plec
cat_cols = [col for col in df.columns if df[col].dtype == 'O'] #sprawdzenie ktore kolumny maja typ litea O czyli sa obiektem
cat_cols
````

````python
#iteracja po kazdej kategorii i zmiana jej typu na category
for col in cat_cols:
  df[col] = df[col].astype('category')
df.info()
````

````python
df.describe().T #wyswietlenie danych statystycznych o calym zbiorze df
````

````python
df.describe(include=['category']).T #wyswietlenie danych statystycznych dot.danych kategorycznych
````

````python
df.sex.value_counts() #sprawdza rozklad zmiennej plec
````

````python
df.charges.plot(kind='hist', bins=30) #wykres histogram dla zmiennej charges
````

````python
df_dummies = pd.get_dummies(df, drop_first=True) #przygotowanie danych do modelu, rozbicie kolumn kategorycznych na liczbowe
df_dummies
````

````python
corr = df_dummies.corr() #badanie korelacji dla poszczegolnych zmiennych
corr
````

````python
df_dummies.corr()['charges'].sort_values(ascending=False) #wyciagniecie kolreacji dla zmiennej charges
````

````python
#przygotowanie bezposrednio do modelu
data = df_dummies.copy() #dane na podstawie ktorych bedziemy budowac model
target = data.pop('charges')
data.head()#dane na podstawie ktorych bedziemy budowac model
````

````python
target.head() #dane ktore bedziemy chcieli przewidywac
````

````python
#podzial danych na zb. treningowy i testowy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)

#wydrukowanie rozmiaru zbiorow danych
print(f'X_trian shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_trian shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
````

````python
#budowanie modelu regresji liniowej
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()   #inskancja klasy
regressor.fit(X_train, y_train)  #dopasowujemy nasz model na danych treningowych

print(f'R2 score: {regressor.score(X_test, y_test):.4f}')  #wydrukowanie wskaznik dopasowania

#wykonanie predykcji i wyswietelenie 10 elementow ktore model przewidzial
y_pred = regressor.predict(X_test)
y_pred[:10]
````

````python
#porownanie wartosci przewidzianych przez model z rzeczywistymi i sprawdzenie poziomu roznicy
y_true = y_test.copy()
predictions = pd.DataFrame(data={'y_true': y_true, 'y_pred': y_pred})
predictions['error'] = predictions['y_true'] -  predictions['y_pred']
predictions.head()
````

###ELIMINACJA WSTECZNA - dobor cech modelu

````python
#uzycie modelu OLS
import statsmodels.api as sm

X_train_ols = X_train.copy() #zmienna ktora pozwoli zobaczyc jak dobierac konkretne dane
# Convert boolean columns to integers
for col in X_train_ols.columns:
  if X_train_ols[col].dtype == 'bool':
    X_train_ols[col] = X_train_ols[col].astype(int)

X_train_ols = X_train_ols.values
X_train_ols = sm.add_constant(X_train_ols)  #dodanie stalej '1' - bibliteka statsmodels tego wymaga
X_train_ols
````

````python
#metoda OLS to regresja liniowa w bibl statsmodel
ols = sm.OLS(endog=y_train, exog=X_train_ols).fit()
predictors = ['const'] + list(X_train.columns)
print(ols.summary(xname=predictors))
````

````python
from os import remove
#usuwamy pojedynczo zmienne o p>t wiekszym od 0.05
#wybieramy zmienne bez sex_male(4)
X_selected = X_train_ols[:, [0,1,2,3,5,6,7,8]]
predictors.remove('sex_male')

#dopsujemy jeszcze raz model
ols = sm.OLS(endog=y_train, exog=X_selected).fit()
print(ols.summary(xname=predictors))
````

````python
#usuwamy kolejna zmienna
X_selected = X_train_ols[:, [0,1,2,3,5,7,8]]
predictors.remove('region_northwest')

#dopsujemy jeszcze raz model
ols = sm.OLS(endog=y_train, exog=X_selected).fit()
print(ols.summary(xname=predictors))
````

````python
#usuwamy kolejna zmienna
X_selected = X_train_ols[:, [0,1,2,3,5,7]]
predictors.remove('region_southwest')

#dopsujemy jeszcze raz model
ols = sm.OLS(endog=y_train, exog=X_selected).fit()
print(ols.summary(xname=predictors))
````

````python
#usuwamy kolejna zmienna
X_selected = X_train_ols[:, [0,1,2,3,5]]
predictors.remove('region_southeast')

#dopsujemy jeszcze raz model
ols = sm.OLS(endog=y_train, exog=X_selected).fit()
print(ols.summary(xname=predictors))
````
